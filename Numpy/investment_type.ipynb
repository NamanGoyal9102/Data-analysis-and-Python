```python
# Investment Type
# Send Feedback
# Given File 'startup_funding.csv'
# Problem Statement :
# There are 4 different type of investments. Find out percentage of amount funded for each investment type.
# Plot a pie chart to visualise.
# Print the investment type and percentage of amount funded with 2 decimal places after rounding off.
# Note :
# Correct spelling of investment types are - "Private Equity", "Seed Funding", "Debt Funding", and "Crowd Funding". Keep an eye for any spelling mistake. You can find this by printing unique values from this column.
# Print the investment type in descending order with respect to the percentage of the amount funded.
# Output Format :
# investmentType1 percent1
# investmentType2 percent2
# investmentType3 percent3
# . . . 


# Open and read data file as specified in the question
# Print the required output in given format
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline


df = pd.read_csv('startup_funding.csv',encoding = 'utf-8')
df.dropna(subset=['InvestmentType'],inplace=True)
# print(df['InvestmentType'].unique())
df[df['InvestmentType']=='SeedFunding']='Seed Funding'
df[df['InvestmentType']=='Crowd funding']='Crowd Funding'
df[df['InvestmentType']=='PrivateEquity']='Private Equity'
# print(df['AmounInUSD'])
df[df['AmountInUSD']=='####']=None
df[df['AmountInUSD']=='Seed Funding']=None
df[df['AmountInUSD']=='Private Equity']=None
df[df['AmountInUSD']=='Crowd Funding']=None
df.dropna(subset=['AmountInUSD'],inplace=True)

def ans(num):
    num=float(num.replace(",",''))
    return int(num)
df['AmountInUSD']=df['AmountInUSD'].apply(ans)

# df['AmountInUSD']=df['AmountInUSD'].apply(int)
# print(df['InvestmentType'].unique())
dict=df.groupby('InvestmentType')['AmountInUSD'].sum()
df = dict.sort_values(ascending = False)
# print(df)
val=np.array(df.values)
key=np.array(df.index)
sum=np.sum(val)
# print(sum)

dict=dict.to_dict()
keys=dict.keys()
values=dict.values()
plt.pie(values,labels=keys,autopct="%.2f%%")
plt.show()
for i in range(len(key)):
    print(key[i],format(val[i]/sum*100,'0.2f'))
    

```
