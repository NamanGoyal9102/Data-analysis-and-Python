```python
## Open and read data file as specified in the question
## Print the required output in given format
import numpy as np
import csv
data=open('terrorismData.csv')
file_set=csv.DictReader(data,skipinitialspace=True)
# print(file_set['State'])
cities={}
woundedcities={}
for row in file_set:
    if row['Country']=='India' and row['City']!='Unknown':
    
        if row['City'] in cities:
            if row['Killed']=='' and row['Wounded']=='':
                continue
            elif row['Killed']=='' and row['Wounded']!='':
                cities[row['City']]=float(row['Wounded'])+cities[row['City']]
            elif row['Killed']!='' and row['Wounded']=='':
                cities[row['City']]=float(row['Killed'])+cities[row['City']]
            
            
            else:
                cities[row['City']]=float(row['Killed'])+float(row['Wounded'])+cities[row['City']]
        else:
            if row['Killed']=='':
                continue
            elif row['Killed']=='' and row['Wounded']!='':
                cities[row['City']]=float(row['Wounded'])
            elif row['Killed']!='' and row['Wounded']=='':
                cities[row['City']]=float(row['Killed'])
            else:
                cities[row['City']]=float(row['Killed'])+float(row['Wounded'])
    else:
        continue

        
sorted_cities=sorted(cities.items(), key=lambda x:x[1])
# THIS SORTED METHOD ALWAYS CONVERTS THE GIVEN DICTIONARY INTO THE LIST OF TUPLES , ALWAYS THIS IS THE CASE THAT THE SORTED METHOD CONVERTS THE
# GIVEN THING INTO A SORTED LIST
for i in range(0,5):
    print(sorted_cities[len(sorted_cities)-i-1][0],int(sorted_cities[len(sorted_cities)-i-1][1]))
    
    
    
# import numpy as np
# import csv
# from collections import Counter

# with open('terrorismData.csv',encoding="utf8") as file_obj :
#     data = csv.DictReader(file_obj,skipinitialspace=True)
    
#     city = []
#     casualty = []
#     for row in data:
#          if row["Country"] == 'India' and row['City'] != 'Unknown':
#                 casualty.append([row["Killed"],row["Wounded"]])
#                 city.append(row["City"])
                
#     np_city = np.array(city)
#     np_casualty = np.array(casualty)
    
#     np_casualty[np_casualty==''] = "0.0"
#     np_casualty = np.array(np_casualty, dtype = float)
#     np_casualty = np.sum(np_casualty, axis=1)
    
#     dic = {}
#     for i in range(len(np_city)):
#         if np_city[i] in dic:
#             dic[np_city[i]] += np_casualty[i]
#         else:
#             dic[np_city[i]] = np_casualty[i]

#     k = Counter(dic) 
#     high = k.most_common(5) 
#     for i in high: 
#         print(i[0],int(i[1])) 
```
